```python
import pandas as pd
import numpy as np
```

```python
#Lectura del archivo JSON
pd.read_json('TelecomX_Data.json')
```

```python
#Creación del DataFrame
datos = pd.read_json('TelecomX_Data.json')
datos.head()
```

```python
#Normalizacion de los datos
dato_customer = pd.json_normalize(datos['customer'])
dato_phone = pd.json_normalize(datos['phone'])
dato_internet = pd.json_normalize(datos['internet'])
dato_account = pd.json_normalize(datos['account'])

datos_normalizados = pd.concat([datos[['customerID', 'Churn']],
                                dato_customer,
                                dato_phone,
                                dato_internet,
                                dato_account], axis=1)

display(datos_normalizados.head(25))
```

```python
#Cambiando el Dtype de la columna Charges.total a float para una mejor lectura
datos_normalizados['Charges.Total'] = datos_normalizados['Charges.Total'].replace(' ', np.nan).astype(float)
datos_normalizados.info()
```

```python
#Creando Selda "Cuenta diarias" para una mejor lectura de la tabla
cuentas_diarias = datos_normalizados['Charges.Monthly'] / 30
datos_normalizados['Charges.Daily'] = cuentas_diarias
datos_normalizados.head(5002)
```

```python
#Pasando los datos 'STR' a valores BOOLEANOS para un mejor manejo
columnas = ['Churn', 'Partner', 'Dependents','PhoneService','MultipleLines','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling']
mapeo = {'Yes': 1, 'No': 0}
datos_normalizados['Churn'] = datos_normalizados['Churn'].map(mapeo)
```

```python
#Visualizando la cantidad de valores NaN en la columna 'Churn'
print("Valores faltantes en 'Churn':", datos_normalizados['Churn'].isnull().sum())
```

```python
#Monstrando ejemplos aleatorios de la columna 'Churn'
datos_normalizados['Churn'].sample()
```

```python
#Transformando los valores NaN a moda (El resultado mas probable) para una lectura de grafico mas comoda
moda_churn = datos_normalizados['Churn'].mode()[0]
datos_normalizados['Churn'] = datos_normalizados['Churn'].fillna(moda_churn)
```

```python
#Cambiando el tipo de datos en la columna 'Churn' a int para manejarlo en la Grafica
datos_normalizados['Churn'] = datos_normalizados['Churn'].astype(np.int64)
```

```python
datos_normalizados['Churn'].sample()
```

```python
import matplotlib.pyplot as plt
```

```python
#Definiendo variables para graficar
#Cantidad
clientes_churn = datos_normalizados['Churn'].value_counts()
clientes_churn
cantidad_total = clientes_churn.sum()
#Variable si o no (ilustrativa )
clientes_churn_labels = ['No dejo', 'Dejo']
```

```python
#Grafico
plt.figure(figsize=(8, 6))
plt.pie(clientes_churn,
        labels=clientes_churn_labels,
        autopct=lambda pct: f'{int(pct / 100 * cantidad_total)} ({pct:.1f}%)',
        startangle=140)
plt.title('Distribución de Clientes', loc='left')
plt.axis('equal')
plt.show()
```

```python
#Creando variales para Genero, Tipo de contrato, metodo de pago, etc
genero = datos_normalizados['gender'].value_counts()
tipo_contrato = datos_normalizados['Contract'].value_counts()
metodo_pago = datos_normalizados['PaymentMethod'].value_counts()
facturan = datos_normalizados['PaperlessBilling'].value_counts()
suscripciones = datos_normalizados[['MultipleLines','PhoneService','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies']].value_counts()
```

```python
#Generando grafica
plt.figure(figsize=(10, 6))
plt.pie(genero, labels=genero.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Género')
plt.axis('equal')
plt.show()
```

```python
# Preparar datos
genero_churn = datos_normalizados.groupby(['gender', 'Churn']).size().unstack()

# Grafico con barras
genero_churn.plot(kind='bar', figsize=(10, 6), color=['skyblue', 'salmon'], edgecolor='black')
plt.title('Distribución de Churn por Género')
plt.xlabel('Género')
plt.ylabel('Cantidad de Clientes')
plt.xticks(rotation=0)
plt.legend(['No Dejó', 'Dejó'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Seleccionar columnas de servicios
servicios = [
    'MultipleLines', 'PhoneService', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Contar cuántos 'Yes' hay en cada servicio (asumiendo que 'Yes' indica que lo tiene)
conteo_servicios = {}
for servicio in servicios:
    if servicio == 'InternetService':
        # Esta columna tiene valores como 'Fiber optic', 'DSL', etc.
        continue  # La graficamos aparte
    conteo_servicios[servicio] = (datos_normalizados[servicio] == 'Yes').sum()

# Convertir a serie y graficar

serie_servicios = pd.Series(conteo_servicios).sort_values(ascending=False)

plt.figure(figsize=(10, 6))
serie_servicios.plot(kind='bar', color='violet', edgecolor='black')
plt.title('Cantidad de Clientes con Cada Servicio Adicional')
plt.xlabel('Servicio')
plt.ylabel('Clientes con el Servicio (Yes)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
